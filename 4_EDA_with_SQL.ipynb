```python
QUERY = """
SELECT ConvertedCompYearly
FROM survey
"""

df = pd.read_sql_query(QUERY,conn)
df
```

```python
df["ConvertedCompYearly"].replace("NA",np.nan,inplace=True)
```

```python
df[["ConvertedCompYearly"]]
```

```python
df1 = df[["ConvertedCompYearly"]].dropna().copy()
```

```python
df1 = df1.astype("int64")
df1.dtypes
```

Find out if outliers exist in the column `ConvertedCompYearly` using a box plot?

```python
df1["ConvertedCompYearly"].plot(kind="box",figsize=(20,10))
```

```python
df1.sort_values(by ="ConvertedCompYearly", ascending=False)
```

Find out the Inter Quartile Range for the column `ConvertedCompYearly`.

```python
Q1 = df1['ConvertedCompYearly'].quantile(0.25)
Q3 = df1['ConvertedCompYearly'].quantile(0.75)
print(Q1," ",Q3)
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.

```python
max = df1['ConvertedCompYearly'].max()
min = df1['ConvertedCompYearly'].min()
print (max)
print (min)
```

Identify how many outliers are there in the `ConvertedCompYearly` column.

```python
((df1['ConvertedCompYearly'] < (Q1 - 1.5 * IQR)) | (df1['ConvertedCompYearly'] > (Q3 + 1.5 * IQR))).sum()
```

As we can see that above there are `2831` outliers.

Create a new dataframe by removing the outliers from the `ConvertedCompYearly` column.

```python
mask = (df1['ConvertedCompYearly'] < (Q1 - 1.5 * IQR)) | (df1['ConvertedCompYearly'] > (Q3 + 1.5 * IQR))
df1[mask] = np.nan
df1['ConvertedCompYearly']
```

```python
df1["ConvertedCompYearly"].max()
```

```python
df1["ConvertedCompYearly"].plot(kind="box",figsize=(20,10))
```
